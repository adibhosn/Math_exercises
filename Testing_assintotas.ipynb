```python
from sympy import *
```

```python
x = symbols('x')
y = symbols('y')
```

```python
f = (2*x**2)/sqrt((x**2)-16)
```

```python
plt = plot(f, (x, -10, 10), show = False)
plt.ylim = [-70, 70]

plt.show()
```

TESTANDO ASSINTOTAS HORIZONTAIS

```python
limit(f,x,oo)
```

```python
limit(f,x,-oo)
```

TESTANDO ASSINTOTA VERTICAL

```python
limit(f,x,4,'+')
```

```python
limit(f,x,-4,'-')
```

ASSÍNTOTA APENAS NO LADO POSITIVO

```python
plt = plot(f, (x, -10, 10), show = False)

# Plotando a reta assíntota vertical em x = 2
plt1 = plot_implicit(Eq(x, 4), (x, 1, 4), (y, -70, 70), show = False, line_color = 'r', adaptive = False)

plt.append(plt1[0])
plt.ylim = [-70, 70]
plt.title = "Assíntota Vertical em x = 4"

plt.show()
```

TESTANDO DIAGONAL

```python
limit(f -2*x, x, oo)
```

```python
limit(f - 2*x, x, -oo)
```

```python

plt = plot(f, (x, -10, 10), show = False)

plt1 = plot(2*x, (x, -10, 10), show = False, line_color = 'brown')

plt.append(plt1[0])
plt.title = 'Assíntota Diagonal'

plt.show()
```
